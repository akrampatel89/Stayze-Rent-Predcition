```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("Train.csv")
data.head()
```

```python
le_account=LabelEncoder()
data['account_type']=le_account.fit_transform(data['account_type'])
le_account.classes_
```

```python
le_loan_enq=LabelEncoder()
data['loan_enq']=le_loan_enq.fit_transform(data['loan_enq'])
le_loan_enq.classes_
```

```python
le_gender=LabelEncoder()
data['gender']=le_gender.fit_transform(data['gender'])
le_gender.classes_
```

```python
df=data.set_index('ID').sort_index(ascending=True)
df.head()
```

```python
df.corr()['cc_cons']
```

```python
cols=df.corr()['cc_cons'][abs(df.corr()['cc_cons'])>0.005].index
cols
```

```python
df1=df[cols]
```

```python
df1.head()
```

```python
X=df1.drop('cc_cons',1)
y=df1['cc_cons']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)*100
```

```python
lr.score(X_val, y_val)*100
```

```python
y_train_pred=lr.predict(X_train)
y_val_pred=lr.predict(X_val)
```

```python
def rmsle(actual_column, predicted_column):
    sum=0.0
    for x,y in zip(actual_column,predicted_column):
        if x<0 or y<0: #check for negative values. 
            continue
        p = np.log(y+1)
        r = np.log(x+1)
        sum = sum + (p - r)**2
    return (sum/len(predicted_column))**0.5
```

```python
rmsle(y_train, y_train_pred)
```

```python
rmsle(y_val, y_val_pred)
```

```python
test=pd.read_csv('Test.csv')
test=test.set_index('ID')
test.head()
```

```python
test['account_type']=le_account.transform(test['account_type'])
test['gender']=le_gender.transform(test['gender'])
test['loan_enq']=le_loan_enq.transform(test['loan_enq'])
```

```python
X_test=test[cols.drop('cc_cons')]
```

```python
X_test.head()
```

```python
y_pred=lr.predict(X_test)
```

```python
test['cc_cons']=y_pred
```

```python
result=pd.DataFrame(test['cc_cons'])
result
```

```python
result.to_csv('submission1.csv',index=True)
```
